## CLASS DEFINTIONS:

```python
class Player:

    def __init__(self,chips=0,bet=0):
        self.chips = 100
        self.bet = bet
    
    def __str__(self):
        return ("Chips = {}\nBet = {}".format(self.chips,self.bet))
    
    def chips_new(self,chips_new):
        while True:
            try:
                int(math.floor(float(chips_new)))
            except:
                print('It is not number.')
                chips_new = input('Input your chips\n')
            else:
                self.chips = int(math.floor(float(chips_new)))
                break
                
    def bet_new(self):
        while True:
            bet_new = input('Input your bet\n')
            try:
                int(math.floor(float(bet_new)))
            except:
                print('It is not number.')
                continue
            else:
                if int(math.floor(float(bet_new))) > self.chips:
                    print('Your bet ({}) can not be higer then your chips ({}).'.format(bet_new,self.chips))
                    continue
                else:
                    self.bet = int(math.floor(float(bet_new)))
                    break
        
    def win_bet(self):
        self.chips += self.bet
    
    def lose_bet(self):
        self.chips -= self.bet
```

```python
class Deck:
    
    def __init__(self,deck=None):
        self.deck = deck if deck else [rank + ' of ' + suit for rank in ranks for suit in suits]
        #self.deck = [] 
        #for suit in suits:
         #   for rank in ranks:
          #      self.deck.append(rank + ' of ' + suit)
        
    def __str__(self):
        deck_str = ''
        for card in self.deck:
            deck_str += '\n' + card
        #return ('Deck {}: {}'.format(len(self.deck),self.deck))
        return ('The DECK has ' + str(len(self.deck)) + ' cards:' + deck_str)
    
    def shuffle(self):
        random.shuffle(self.deck)
        #print(self.deck)
    
    def deal(self):
        #card = random.choice(self.deck)
        single_card = self.deck.pop()
        return single_card
```

```python
class Hand:
    
    def __init__(self, cards=None, value=None, aces=None):
        self.cards = []
        self.value = 0
        self.aces = 0
    
    def __str__(self):
        return ('HAND: {}\n\tValue: {}\n\tAces: {}'.format(self.cards,self.value,self.aces))
    
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[str(card)[:str(card).index('of')-1]]
        if str(card)[:str(card).index('of')-1] == 'Ace':
            self.aces += 1  # add to self.aces
        #print('Cards: {} {} {}'.format(self.cards,self.value,self.aces))
    
    def adjust_for_ace(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

## FUNCTION DEFINITIONS:

```python
def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
```

```python
def hit_or_stand(deck,hand):
    while True:
        hit_stand = input("Would you like to Hit or Stand? Enter 'h' or 's' ")
        
        if hit_stand[0].lower() == 'h':
            hit(deck,hand)
            return hit_stand
        elif hit_stand[0].lower() == 's':
            print("Player stands. DEALER is playing.")
            return hit_stand
            break
        else:
            print("Sorry, please try again.")
            continue
        break
```

```python
def show_some(hend_p,hend_d):
    print("\nDealer's Hand:")
    print(" <card hidden>")
    print('',hend_d.cards[1])  
    print("\nPlayer's Hand:", *hend_p.cards, sep='\n ')
    
def show_all(hend_p,hend_d):
    print("\nDealer's Hand:", *hend_d.cards, sep='\n ')
    print("Dealer's Hand =",hend_d.value)
    print("\nPlayer's Hand:", *hend_p.cards, sep='\n ')
    print("Player's Hand =",hend_p.value)
```

```python
def score(hend):
    if hend.value == 21:
        print('\nValue = {}'.format(hend.value))
        print('You WIN!!!\nGAME OVER')
        x = 1
    elif hend.value > 21:
        print('\nValue = {}'.format(hend.value))
        print('You LOST!!!\nGAME OVER')
        x = -1
    else:
        print('\nValue = {}'.format(hend.value))
        hit_stand = hit_or_stand(deck,hend)
        if hit_stand == 's':
            x = 2
        else:
            x = 0
    return x
```
