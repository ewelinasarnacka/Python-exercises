## Ecercise 32 - Hangman

```python
import random
```

```python
def random_word():
    with open("30_sowpods.txt") as word:
        words = list(word)
    word = random.choice(words).strip()
    print(word,"\n")
    return word

def input_letter():
    letter = input("\n\nGuess your letter:").upper()
    return letter

def check_input(letter,letter_list,word):
    if letter in letter_list:
        print("You've already asked about this letter")
        m = 0
    else:
        print("New letter")
        m = 1
        
    if letter not in word.upper() and m == 1:
        print("Incorrect")
        incorrect = 1
    elif letter not in word.upper() and m == 0:
        print("Incorrect")
        incorrect = 0
    else:
        print("Correct")
        incorrect = 0
    return m, incorrect

def print_letter(letter_list,word):
    to_write = list()
    for i in range(1,len(word)+1):
        if word[i-1] in letter_list:
            to_write.append(word[i-1])
        else:
            to_write.append("_")
    seperator = ' '
    to_write = seperator.join(to_write)
    print("\t", to_write)
    return to_write

def drawing(incorrect):
    drawing_top = " +=====+\n |   |"
    drawing_middle = {0:" |\n |\n |",
                      1:" |   O\n |\n |",
                      2:" |   O\n |   |",
                      3:" |   O\n |  /|",
                      4:" |   O\n |  /|\\",
                      5:" |   O\n |  /|\\\n |  /",
                      6:" |   O\n |  /|\\\n |  / \\"
                      }
    drawing_botton = " |\n======="
    print(drawing_top)
    print(drawing_middle[min(incorrect,6)])
    print(drawing_botton)
    
```

```python
def main():
    print("Welcome to Hangman!\nYou have 6 attemps.")
    word = random_word()
    letter_list = list()
    to_write_l = list()
    i = 1
    n = 1
    inc = 0
    while len(to_write_l) != len(word) and inc < 6:
        letter = input_letter()
        m_incorrect = check_input(letter,letter_list,word)
        m = m_incorrect[0]
        incorrect = m_incorrect[1]
        letter_list.append(letter)
        to_write = print_letter(letter_list,word)
        to_write_l = to_write.replace(" ","").replace("_","")
        i += 1
        n += m
        inc += incorrect
        drawing(inc)
        print("You have", 6 - inc, "more attempts.")
        print("You made", inc, "mistakes.")
    if len(to_write_l) == len(word):
        print("Success!!!")
        print("You guessed in", n-1, "steps ( min steps =", len(set(to_write_l)),").")
    elif inc >= 6:
        print("You lost!!!")
    again = input("Do you play again (Y/N)?")
    if again == "Y" or again == "y":
        main()
    else:
        print("Goodbay!")

main()
```
