## Exercise 23 - File Overlap

https://www.practicepython.org/exercise/2014/12/14/23-file-overlap.html

```python
nump_file = open("23_primenumbers.txt", "r")
numprime = nump_file.read().splitlines()
print("Prime numbers:\n", numprime)

numh_file = open("23_happynumbers.txt", "r")
numhappy = numh_file.read().splitlines()
print("Happy numbers:\n", numhappy)

overlap_a = set(numprime) - (set(numprime) - set(numhappy))
print("Overlapin numbers\n",overlap_a)

overlap_b = [n for n in numprime if n in numhappy]
print("Overlapin numbers\n",overlap_b)

check = overlap_a - set(overlap_b)
print("check:",check)
```

```python
def to_list(file):
    num_list = []
    with open(file) as f:
        line = f.readline()
        while line:
            num_list.append(int(line))
            line = f.readline()
    print("Numbers in file", file, ":", num_list)
    return num_list

primeslist = to_list("23_primenumbers.txt")
happylist = to_list("23_happynumbers.txt")

overlap = []
for n in happylist:
    if n in primeslist:
        overlap.append(n)
    #n += 1
print("Overlaping numbers:", overlap)

overlap = [n for n in primeslist if n in happylist]
print("Overlaping numbers:", overlap)
```
