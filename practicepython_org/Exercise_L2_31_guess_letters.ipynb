## Exercise 31 - Guess Letters

https://www.practicepython.org/exercise/2017/01/02/31-guess-letters.html

```python
word = "EVA"
print(word, "\n")

def input_letter():
    letter = input("Guess your letter:\n").upper()
    return letter

def check_input(letter,letter_list,word):
    if letter in letter_list:
        print("You've already asked about this letter")
    else:
        print("New letter")
        
    if letter not in word.upper():
        print("Incorrect")
    else:
        print("Correct")

def print_letter(letter_list,word):
    to_write = list()
    for i in range(1,len(word)+1):
        if word[i-1] in letter_list:
            to_write.append(word[i-1])
        else:
            to_write.append("_")
    seperator = ' '
    to_write = seperator.join(to_write)
    print("\t", to_write)
    return to_write


print("Welcome to Hangman!\n")
def main():
    letter_list = list()
    to_write_l = list()
    i = 0
    while len(to_write_l) != len(word):
        letter = input_letter()
        check_input(letter,letter_list,word)
        letter_list.append(letter)
        to_write = print_letter(letter_list,word)
        to_write_l = to_write.replace(" ","").replace("_","")
        print(to_write_l)
        i += 1
    print("Success!!!")
    print("You guessed in", i, "steps ( min steps =", len(set(to_write_l)),").")

main()
```
